```python
class light:
  def __init__(self,room,st):
    self.room = room
    self.status = st

  def change_status(self):
    if self.status == "on":
      status = "off"
    else:
        status = "on"

  def show(self):
    print(f"{self.room} light is turned {self.status}")





l1 = light("Living Room","on")
l2 = light("Bedroom light","off")

l1.show()
l2.show()

l1.change_status()
l2.change_status()
l1.show()
l2.show()


```

# Practice task-2

```python
class Staff:
  def __init__(self,name,staff_id,department):
    self.name = name
    self.staff_id = staff_id
    self.department = department

  def display_info(self):
    print(f"Name: {self.name}, Staff ID: {self.staff_id}, Department: {self.department}")

class Teacher(Staff):
  def __init__(self, name, staff_id, department,courses,salary):
    super().__init__(name, staff_id, department)
    self.courses = courses
    self.salary = salary

  def teaching(self):
    print(f"Teacher {self.name} is teaching {self.courses} to the class")

  def display_info(self):
    super().display_info()
    print(f"Courses: {self.courses}, Salary: ${self.salary}")


class Admin_staff(Staff):
  def __init__(self, name, staff_id, department,role,working_hours):
    super().__init__(name, staff_id, department)
    self.role = role
    self.working_hours = working_hours

  def performing_task(self):
    print(f"{self.name} from Admin staff is performing his/her tasks")

  def display_info(self):
    super().display_info()
    print(f"Role: {self.role}, Working Hours: {self.working_hours} hours/week")

class Research_Assistant(Staff):
  def __init__(self, name, staff_id, department,research_topic,stipend):
    super().__init__(name, staff_id, department)
    self.research_topic = research_topic
    self.stipend = stipend

  def research(self):
    print(f"{self.name} is researching on {self.research_topic}")

  def display_info(self):
    super().display_info()
    print(f"Research Topic: {self.research_topic}, Stipend: ${self.stipend}")





teacher1 = Teacher("Alice Smith", "T101", "Computer Science", "Programming, Data Structures", 60000)
admin1 = Admin_staff("Bob Johnson", "A201", "Admissions", "Registrar", 40)
researcher1 = Research_Assistant("Charlie Brown", "R301", "Physics", "Quantum Computing", 25000)

print("\n--- Teacher Info ---")
teacher1.display_info()
teacher1.teaching()

print("\n--- Admin Staff Info ---")
admin1.display_info()
admin1.performing_task()

print("\n--- Research Assistant Info ---")
researcher1.display_info()
researcher1.research()
```

```python
class Account:
  def __init__(self,balance):
    self.__balance = balance

  def deposit(self,amount):
    self.__balance += amount

  def withdraw(self,amount):
    if amount <= self.__balance:
      self.__balance -= amount
    else:
      print("Insufficient funds")

  def get_balance(self):
    return self.__balance


a1 = Account(1000)
a1.deposit(500)
a1.withdraw(200)
print(a1.get_balance())

a2 = Account(2000)
a2.deposit(1000)
a2.withdraw(3000)
print(a2.get_balance())

```

```python
import math

class point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def display(self):
    print(f"({self.x},{self.y})")

class Shape:

  def Area(self):
    pass

  def perimeter(self):
    pass

  def draw(self):
    pass

class Line(Shape):
  def __init__(self,p1,p2):
    self.start = p1
    self.end = p2

  def Area(self):
    return 0

  def perimeter(self):
    return math.dist((self.start.x,self.start.y),(self.end.x,self.end.y))

  def draw(self):
    print("start:",end="")
    self.start.display()
    print("end:",end="")
    self.end.display()


class Triangle(Shape):
  def __init__(self,p1,p2,p3):
      self.p1 = p1
      self.p2 = p2
      self.p3 = p3


  def perimeter(self):
    return (math.dist((self.p1.x,self.p1.y),(self.p2.x,self.p2.y))+
           math.dist((self.p2.x,self.p2.y),(self.p3.x,self.p3.y))+
           math.dist((self.p3.x,self.p3.y),(self.p1.x,self.p1.y)))

  def Area(self):
        # Assuming a, b, c are point objects
        # This formula seems to be for triangle area given coordinates of vertices.
        # The current implementation uses self.a, self.b, self.c which are not defined.
        # It should use self.p1, self.p2, self.p3
        return abs(
            (self.p1.x*(self.p2.y-self.p3.y) +
             self.p2.x*(self.p3.y-self.p1.y) +
             self.p3.x*(self.p1.y-self.p2.y)) / 2)

  def draw(self):
    print("Triangle at points:")
    self.p1.display()
    self.p2.display()
    self.p3.display()

class Reactangle(Shape):
  def __init__(self,top_left,width,height):
    self.top_left = top_left
    self.width = width
    self.height = height

  def perimeter(self):
    return 2*(self.width+self.height)

  def Area(self):
    return self.width*self.height

  def draw(self):
    print(f"Reactangle has width = {self.width}, height = {self.height} and top left corner at:")
    self.top_left.display()


class Square(Reactangle):
  def __init__(self, top_left, width):
    super().__init__(top_left, width, width)

  def draw(self):
    print(f"Square has side = {self.width} and top left corner at:")
    self.top_left.display()


class Circle(Shape):
  def __init__(self,center,radius):
    self.center = center
    self.radius = radius

  def perimeter(self):
      return 2*math.pi*self.radius

  def Area(self):
    return math.pi*self.radius**2

  def draw(self):
    print(f"Circle has radius = {self.radius} and center at:")
    self.center.display()


class Pentagon(Shape):
  def __init__(self,p1,p2,p3,p4,p5):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4
    self.p5 = p5

  def perimeter(self):
    return (
        math.dist((self.p1.x, self.p1.y), (self.p2.x, self.p2.y)) +
        math.dist((self.p2.x, self.p2.y), (self.p3.x, self.p3.y)) +
        math.dist((self.p3.x, self.p3.y), (self.p4.x, self.p4.y)) +
        math.dist((self.p4.x, self.p4.y), (self.p5.x, self.p5.y)) +
        math.dist((self.p5.x, self.p5.y), (self.p1.x, self.p1.y))
    )

  def Area(self):
    points = [(self.p1.x, self.p1.y),
              (self.p2.x, self.p2.y),
              (self.p3.x, self.p3.y),
              (self.p4.x, self.p4.y),
              (self.p5.x, self.p5.y)]

    n = len(points)
    area = 0.0

    for i in range(n):
        j = (i + 1) % n
        area += points[i][0] * points[j][1]
        area -= points[j][0] * points[i][1]

    return abs(area / 2.0)

  def draw(self):
    print("Pentagon at points:")
    self.p1.display()
    self.p2.display()
    self.p3.display()
    self.p4.display()
    self.p5.display()


class Quadilateral(Shape):
  def __init__(self,p1,p2,p3,p4):
    self.p1 = p1
    self.p2 = p2
    self.p3 = p3
    self.p4 = p4

  def perimeter(self):
    return (
        math.dist((self.p1.x, self.p1.y), (self.p2.x, self.p2.y)) +
        math.dist((self.p2.x, self.p2.y), (self.p3.x, self.p3.y)) +
        math.dist((self.p3.x, self.p3.y), (self.p4.x, self.p4.y)) +
        math.dist((self.p4.x, self.p4.y), (self.p1.x, self.p1.y))
    )

  def Area(self):
    points = [(self.p1.x, self.p1.y),
              (self.p2.x, self.p2.y),
              (self.p3.x, self.p3.y),
              (self.p4.x, self.p4.y)]

    n = len(points)
    area = 0.0

    for i in range(n):
        j = (i + 1) % n
        area += points[i][0] * points[j][1]
        area -= points[j][0] * points[i][1]

    return abs(area / 2.0)

  def draw(self):
    print("Quadilateral at points:")
    self.p1.display()
    self.p2.display()
    self.p3.display()
    self.p4.display()


p_line1 = point(1, 2)
p_line2 = point(4, 2)
line1 = Line(p_line1, p_line2)
print("--- Line Test ---")
line1.draw()
print(f"Perimeter (Length): {line1.perimeter():.2f}")
print(f"Area: {line1.Area()}")


p_tri1 = point(0, 0)
p_tri2 = point(3, 0)
p_tri3 = point(1.5, 4)
triangle1 = Triangle(p_tri1, p_tri2, p_tri3)
print("\n--- Triangle Test ---")
triangle1.draw()
print(f"Perimeter: {triangle1.perimeter():.2f}")
print(f"Area: {triangle1.Area():.2f}")


p_rect_top_left = point(0, 5)
rectangle1 = Reactangle(p_rect_top_left, 6, 3)
print("\n--- Rectangle Test ---")
rectangle1.draw()
print(f"Perimeter: {rectangle1.perimeter():.2f}")
print(f"Area: {rectangle1.Area():.2f}")


p_square_top_left = point(1, 1)
square1 = Square(p_square_top_left, 4)
print("\n--- Square Test ---")
square1.draw()
print(f"Perimeter: {square1.perimeter():.2f}")
print(f"Area: {square1.Area():.2f}")


p_circle_center = point(5, 5)
circle1 = Circle(p_circle_center, 3)
print("\n--- Circle Test ---")
circle1.draw()
print(f"Perimeter (Circumference): {circle1.perimeter():.2f}")
print(f"Area: {circle1.Area():.2f}")


p_pent1 = point(0, 0)
p_pent2 = point(2, 0)
p_pent3 = point(3, 2)
p_pent4 = point(1, 4)
p_pent5 = point(-1, 2)
pentagon1 = Pentagon(p_pent1, p_pent2, p_pent3, p_pent4, p_pent5)
print("\n--- Pentagon Test ---")
pentagon1.draw()
print(f"Perimeter: {pentagon1.perimeter():.2f}")
print(f"Area: {pentagon1.Area():.2f}")


p_quad1 = point(0, 0)
p_quad2 = point(5, 0)
p_quad3 = point(6, 3)
p_quad4 = point(1, 4)
quadilateral1 = Quadilateral(p_quad1, p_quad2, p_quad3, p_quad4)
print("\n--- Quadilateral Test ---")
quadilateral1.draw()
print(f"Perimeter: {quadilateral1.perimeter():.2f}")
print(f"Area: {quadilateral1.Area():.2f}")



```

```python
class Array:
  def __init__(self,size):
    self.arr = []*size

  def insert(self,n):
    self.arr.append(n)

  def delete(self,idx):
    self.arr.pop(idx)


  def search(self,n):
    for i in range(len(self.arr)):
      if self.arr[i] == n:
        return i
    print(f"{n} is not in the array")

  def display(self):
    print(self.arr)

class Stack:
  def __init__(self,size):
    self.arr = []*size

  def insert(self,n):
    self.arr.append(n)

  def delete(self):
    self.arr.pop(-1)

  def search(self,n):
    for i in range(len(self.arr)):
      if self.arr[i] == n:
        return i

    print(f"{n} is not in the stack")

  def display(self):
    print(self.arr)


class Queue:
  def __init__(self,size):
    self.arr = []*size


  def insert(self,n):
    self.arr.append(n)

  def delete(self):
    self.arr.pop(0)

  def display(self):
    print(self.arr)


print("--- Array Test ---")
my_array = Array(5)
my_array.insert(10)
my_array.insert(20)
my_array.insert(30)
my_array.display()
print(f"Search for 20: Index {my_array.search(20)}")
my_array.delete(1)
my_array.display()
print(f"Search for 50: {my_array.search(50)}")


print("\n--- Stack Test ---")
my_stack = Stack(5)
my_stack.insert('A')
my_stack.insert('B')
my_stack.insert('C')
my_stack.display()
my_stack.delete()
my_stack.display()
print(f"Search for 'A': Index {my_stack.search('A')}")
my_stack.delete()
my_stack.delete()
my_stack.display()


print("\n--- Queue Test ---")
my_queue = Queue(5)
my_queue.insert('X')
my_queue.insert('Y')
my_queue.insert('Z')
my_queue.display()
my_queue.delete()
my_queue.display()
my_queue.insert('W')
my_queue.display()
my_queue.delete()
my_queue.delete()
my_queue.delete()
my_queue.display()




```
