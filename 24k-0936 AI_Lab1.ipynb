#Task-1

```python
def fibo(num):
  if num <= 1:
    return num

  return fibo(num-1)+fibo(num-2)

n = int(input("enter the number:"))
for i in range(n):
  print(fibo(i))
```

#Task-2

```python
import math
def isprime(n):
  if n<2:
    return False

  for i in range(2,int(math.sqrt(n))+1):
    if n%i == 0:
      return False

  return True

l = []
n = int(input("enter number:"))

for i in range(n):
  num = int(input("enter element:"))
  l.append(num)
  if isprime(num):
    print(f"{num} is prime")
  else:
    print(f"{num} is not prime")

print("list:")
for i in l:
  print(i)
```

# Task-3

```python
def isAnagram(str1,str2):
  freq = {}

  for s in str1:
      freq[s] = 0


  for s in str1:
      freq[s]+=1


  for s in str2:
    if s in freq.keys():
      freq[s]-=1
    else:
      return False

  for v in freq.values():
    if v != 0:
      return False

  return True


str1 = input("enter string 1:")
str2 = input("enter string 2:")

print(str1,str2)

isAnagram(str1,str2)
```
