```python
class WarehouseEnvironment:
    def __init__(self, start_position, goal_position, path_length):
        self.current_position = start_position
        self.goal_position = goal_position
        self.path_length = path_length

    def get_current_position(self):
        return self.current_position

    def set_current_position(self, new_position):
        if 0 <= new_position < self.path_length:
            self.current_position = new_position
        else:
            print(f"invaid motion! Position remains {self.current_position}")

    def get_goal_position(self):
        return self.goal_position

    def is_goal_reached(self):
        return self.current_position == self.goal_position

class DeliveryRobot:
    def __init__(self, environment):
        self.environment = environment
        self.steps_taken = 0
        self.positions_visited = [environment.get_current_position()]

    def decide_and_act(self):
        current_pos = self.environment.get_current_position()
        goal_pos = self.environment.get_goal_position()

        if current_pos < goal_pos:
            new_pos = current_pos + 1
            self.environment.set_current_position(new_pos)
            self.steps_taken += 1
            self.positions_visited.append(new_pos)
            return True
        else:
            return False

    def run_simulation(self):
        print(f"Robot simulation started. Initial position: {self.environment.get_current_position()}, Goal: {self.environment.get_goal_position()}\n")

        while not self.environment.is_goal_reached():
            action_performed = self.decide_and_act()
            if not action_performed:
                break

        print("Simulation Complete")
        print(f"Robot successfully reached the goal at position: {self.environment.get_current_position()}")
        print(f"Total steps taken: {self.steps_taken}")
        print(f"Sequence of positions visited: {self.positions_visited}")


warehouse = WarehouseEnvironment(0, 15, 16)
robot = DeliveryRobot(warehouse)


robot.run_simulation()
```

```python
class DeliveryOption:
    def __init__(self, name, time, cost):
        self.name = name
        self.time = time
        self.cost = cost

    def __str__(self):
        return f"Option {self.name}: Time = {self.time} days, Cost = ${self.cost:.2f}"


class Customer:
    def __init__(self, name, time_weight, cost_weight):
        self.name = name
        self.time_weight = time_weight
        self.cost_weight = cost_weight

    def __str__(self):
        return f"Customer {self.name}: Time Importance = {self.time_weight}, Cost Importance = {self.cost_weight}"


class ShoppingAssistant:
    def __init__(self, customer, options):
        self.customer = customer
        self.options = options

    def _normalize_attributes(self, options):
        all_times = [option.time for option in options]
        all_costs = [option.cost for option in options]

        min_time = min(all_times)
        max_time = max(all_times)
        min_cost = min(all_costs)
        max_cost = max(all_costs)

        normalized_data = {}
        for option in options:
            normalized_time = (max_time - option.time) / (max_time - min_time) if (max_time - min_time) != 0 else 1.0
            normalized_cost = (max_cost - option.cost) / (max_cost - min_cost) if (max_cost - min_cost) != 0 else 1.0

            normalized_data[option.name] = {
                'normalized_time': normalized_time,
                'normalized_cost': normalized_cost
            }
        return normalized_data


    def calculate_utility(self, option, normalized_data):
        option_norm_data = normalized_data[option.name]
        utility = (self.customer.time_weight * option_norm_data['normalized_time']) + \
                  (self.customer.cost_weight * option_norm_data['normalized_cost'])
        return utility

    def recommend_option(self):
        print(f"\nRecommending for {self.customer.name}")
        print(f"Customer preferences: Time Importance = {self.customer.time_weight}, Cost Importance = {self.customer.cost_weight}")

        if not self.options:
            return "No delivery options available."

        normalized_attributes = self._normalize_attributes(self.options)

        best_option = None
        max_utility = -1
        utility_scores = {}

        for option in self.options:
            utility = self.calculate_utility(option, normalized_attributes)
            utility_scores[option.name] = utility
            print(f"{option} -> Utility Score: {utility:.2f}")

            if utility > max_utility:
                max_utility = utility
                best_option = option

        if best_option:
            print(f"\nRecommendation for {self.customer.name}: {best_option.name} (Utility: {max_utility:.2f})")
            return best_option
        else:
            return "Could not recommend an option."



option_a = DeliveryOption("A", time=2, cost=15.00)
option_b = DeliveryOption("B", time=5, cost=5.00)

delivery_options = [option_a, option_b]

print("Available Delivery Options:")
for opt in delivery_options:
    print(opt)


customer1 = Customer("Alice", time_weight=0.6, cost_weight=0.4)
customer2 = Customer("Bob", time_weight=0.4, cost_weight=0.6)
customer3 = Customer("Charlie", time_weight=0.5, cost_weight=0.5)

assistant1 = ShoppingAssistant(customer1, delivery_options)
assistant1.recommend_option()

assistant2 = ShoppingAssistant(customer2, delivery_options)
assistant2.recommend_option()

assistant3 = ShoppingAssistant(customer3, delivery_options)
assistant3.recommend_option()
```
